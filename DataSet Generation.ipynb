```python
pricing = {
    'Not Rented': {'base_fare': 0, 'rate_per_hour': 0, 'rate_per_km': 0},
    'Hatchback': {'base_fare': 200, 'rate_per_hour': 50, 'rate_per_km': 8},
    'Sedan': {'base_fare': 300, 'rate_per_hour': 70, 'rate_per_km': 10},
    'SUV': {'base_fare': 400, 'rate_per_hour': 90, 'rate_per_km': 12},
    'Luxury': {'base_fare': 800, 'rate_per_hour': 150, 'rate_per_km': 20},
}
```

```python
# ren_car = ren_car.drop(['code','price'], axis=1)
ren_car = ren_car.rename(columns={'totalDistance':'Car_total_distance'})
ren_car
```

```python
def calculate_rent(row):
    car = row['carType']
    
    # Handle cases where carType is not in pricing dictionary
    if car not in pricing or pd.isna(car) or car == 'Not Rented':
        return 0.0  # No charge for Not Rented cars
    
    base_fare = pricing[car]['base_fare']
    rate_per_hour = pricing[car]['rate_per_hour']
    rate_per_km = pricing[car]['rate_per_km']

    # Handle NaN values by replacing with 0
    rental_duration = row['rentalDuration'] if not pd.isna(row['rentalDuration']) else 0
    distance = row['Car_total_distance'] if not pd.isna(row['Car_total_distance']) else 0

    # Calculate total price
    total_price = (
        base_fare +
        (rental_duration * rate_per_hour) +
        (distance * rate_per_km) +
        np.random.randint(10, 51)  # Random charge between 10 and 50
    )
    
    return total_price

# Apply function
ren_car['total_rent_price'] = ren_car.apply(calculate_rent, axis=1)
```
